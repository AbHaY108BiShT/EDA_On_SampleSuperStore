```python
plt.scatter(df["Sales"] , df["Profit"])
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.title("Sales vs Profit")
sns.set_style("darkgrid")
plt.show()
```

<Font color="gray">At lower Sales Profit is either more or less but at higher Sales Profit is high</Font>

```python
profit_to_sales_ratio = df.groupby("State").sum()[["Sales" , "Profit"]]
profit_to_sales_ratio["PTS Ratio"] = profit_to_sales_ratio["Profit"] / profit_to_sales_ratio["Sales"]
profit_to_sales_ratio.sort_values(by="PTS Ratio" ,ascending = False , inplace = True)
print("Top 5 most profit making states")
print()
print(profit_to_sales_ratio.head(5))
print("----"*14)
print("Least 5 profit making companies")
print()
print(profit_to_sales_ratio.tail(5))
```

<Font color="gray">
    <ul>
        <li>Ohio has the least profit to the sales ratio</li>
        <li>District of Colombia has the highest profit to the sales ratio</li>
    </ul>
</Font>

## Analysis on the basis of Quantity

```python
print("Top 5 State on the basis of quantity of products sold")
quantity_ana  = df.groupby("State").count()[["Quantity"]]
quantity_ana.sort_values(by="Quantity" , ascending = False , inplace = True)
print(quantity_ana.head(5))
print()
print("---"*12)
print()
print("Last 5 State on the basis of quantity of products sold")
print()
print(quantity_ana.tail(5))
```

```python
states_sales = df.groupby("State").sum()[['Sales']]
states_sales.sort_values(by="Sales" , ascending = False , inplace = True)
state_regions = df[["State" , "Region"]].drop_duplicates()
states_sales = pd.merge(states_sales ,state_regions , on =["State"] , how = "left")
states_sales.head(5)
```

```python
plt.figure(figsize = (15 , 5))
sns.barplot(x = "State" , y = "Sales" , data = states_sales , hue ="Region" , dodge = False)
plt.xticks(rotation = 90)
sns.set_style("darkgrid")
plt.show()
```

<Font color="gray">
    <ul>
        <li>California has the highest sales in the West region</li>
        <li>New York has the 2nd highest sales in the East region</li>
        <li>North Dekota has the lowest sales</li>
    </ul>
</Font>
